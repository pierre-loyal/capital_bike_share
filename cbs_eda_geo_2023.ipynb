## Create Season Column

```python
def get_season(date):
    if date.month in [12, 1, 2]:
        return 'winter'
    elif date.month in [3, 4, 5]:
        return 'spring'
    elif date.month in [6, 7, 8]:
        return 'summer'
    elif date.month in [9, 10, 11]:
        return 'autumn'

df_2023['season'] = df_2023['started_at'].apply(get_season)

print(df_2023.head())
```

## Defining day_type

```python
bank_holiday = ['2023-01-02', '2023-01-16', '2023-02-20', '2023-05-29', '2023-06-19', '2023-07-04', '2023-09-04', '2023-10-09', '2023-11-10', '2023-11-23', '2023-12-25']  
school_holidays = [('2023-02-18', '2023-02-24'), ('2023-03-15', '2023-03-21'), ('2023-07-01', '2023-08-27'),('2023-11-22', '2023-11-26'), ('2023-12-22', '2023-12-31')]

df_2023['day_type'] = 'regular'

bank_holiday_dates = pd.to_datetime(bank_holiday).date  
df_2023.loc[df_2023['started_at'].dt.date.isin(bank_holiday_dates), 'day_type'] = 'bank_holiday'

for start, end in school_holidays:
    df_2023.loc[df_2023['started_at'].dt.date.between(pd.to_datetime(start).date(), pd.to_datetime(end).date()), 'day_type'] = 'school_holidays'

```

## Daily Mean Number of Rides per Hour by Season

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, date, time

df_2023['hour'] = df_2023['started_at'].dt.hour
df_2023['date'] = df_2023['started_at'].dt.date  



hourly_counts = df_2023.groupby(['season', 'date', 'hour']).size().reset_index(name='counts')


daily_mean_rides = hourly_counts.groupby(['season', 'hour'])['counts'].mean().reset_index(name='daily_mean_counts')
daily_mean_rides_per_hour = daily_mean_rides.pivot(index='hour', columns='season', values='daily_mean_counts')


plt.figure(figsize=(12, 6))
for season in daily_mean_rides_per_hour.columns:
    plt.plot(daily_mean_rides_per_hour.index, daily_mean_rides_per_hour[season], label=season)

plt.title('Daily Mean Number of Rides per Hour by Season - 2023')
plt.xlabel('Hour of the Day')
plt.ylabel('Daily Mean Number of Rides')
plt.xticks(range(0, 24)) 
plt.legend()
plt.grid(True)
plt.show()
```

## Daily Mean Number of Rides per Hour by Day Type

```python
# Extract hour and date from 'started_at'
df_2023['hour'] = df_2023['started_at'].dt.hour
df_2023['date'] = df_2023['started_at'].dt.date  # Extract date for daily grouping


# Group by 'season', 'date', and 'hour', then count rides per hour for each season and date
hourly_counts = df_2023.groupby(['day_type', 'date', 'hour']).size().reset_index(name='counts')

# Calculate the mean number of rides per hour for each season, across all days
daily_mean_rides = hourly_counts.groupby(['day_type', 'hour'])['counts'].mean().reset_index(name='daily_mean_counts')
daily_mean_rides_per_hour = daily_mean_rides.pivot(index='hour', columns='day_type', values='daily_mean_counts')

# Plot
plt.figure(figsize=(12, 6))
for day_type in daily_mean_rides_per_hour.columns:
    plt.plot(daily_mean_rides_per_hour.index, daily_mean_rides_per_hour[day_type], label=day_type)

plt.title('Daily Mean Number of Rides per Hour by Day Type - 2023')
plt.xlabel('Hour of the Day')
plt.ylabel('Daily Mean Number of Rides')
plt.xticks(range(0, 24))  # Ensure x-axis ticks represent each hour of the day
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_2023 is already filtered to include only the desired year
df_2023['started_at'] = pd.to_datetime(df_2023['started_at'])
df_2023['date'] = df_2023['started_at'].dt.date

# Filter for morning and evening rush hours
morning_rush = df_2023[((df_2023['started_at'].dt.hour == 7) & (df_2023['started_at'].dt.minute >= 30)) |
                        ((df_2023['started_at'].dt.hour == 8) & (df_2023['started_at'].dt.minute <= 30))]

evening_rush = df_2023[((df_2023['started_at'].dt.hour == 16) & (df_2023['started_at'].dt.minute >= 30)) |
                        ((df_2023['started_at'].dt.hour == 17) & (df_2023['started_at'].dt.minute <= 30))]

# Function to calculate daily mean number of rides per station
def daily_mean_rides(data):
    # Group by date and station, count rides, then calculate daily mean
    daily_counts = data.groupby(['date', 'start_station_name']).size().reset_index(name='count')
    daily_mean = daily_counts.groupby('start_station_name')['count'].mean().reset_index(name='daily_mean_count')
    return daily_mean.sort_values(by='daily_mean_count', ascending=False)

# Calculate daily mean number of rides for morning and evening rush hours
morning_mean_rides = daily_mean_rides(morning_rush)
evening_mean_rides = daily_mean_rides(evening_rush)

# Plotting
top_n_stations = 10  # Adjust as needed

# Morning rush hour
plt.figure(figsize=(10, 6))
plt.barh(morning_mean_rides.head(top_n_stations)['start_station_name'], morning_mean_rides.head(top_n_stations)['daily_mean_count'])
plt.xlabel('Daily Mean Number of Rides')
plt.ylabel('Station Name')
plt.title('Top Stations During Morning Rush Hour (7:30-8:30) - Daily Mean')
plt.gca().invert_yaxis()  # Highest counts at the top
plt.show()

# Evening rush hour
plt.figure(figsize=(10, 6))
plt.barh(evening_mean_rides.head(top_n_stations)['start_station_name'], evening_mean_rides.head(top_n_stations)['daily_mean_count'])
plt.xlabel('Daily Mean Number of Rides')
plt.ylabel('Station Name')
plt.title('Top Stations During Evening Rush Hour (16:30-17:30) - Daily Mean')
plt.gca().invert_yaxis()  # Highest counts at the top
plt.show()
```
